```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
links = []
for j in range(1,153):
        link = 'https://www.taiwannews.com.tw/en/search?keyword=trade%20war&page='+str(j)
        print(link)
        links.append(link)
```

```python
news=[]
for link in links:
    res = requests.get(link)
    soup = BeautifulSoup(res.text,'html5lib')
    rows=soup.find_all('header', 'entry-header')
    for row in rows:
        r=row.find_all('a')
        for i in r:
            if 'href' in i.attrs:
                news.append('https://www.taiwannews.com.tw'+i['href'])
```

```python
nn=[]
for i in news:
    new = {}
    res = requests.get(i)
    soup = BeautifulSoup(res.text, 'html5lib')
    rows = soup.find_all('div', 'no-gutter')
    new['發稿日']=rows[0].find('div','article-date').text.strip()
    new['新聞標題']=rows[0].find('h1','article-title').text.strip()
    try :
        article_author=rows[0].find('div','article-author').a['title']
    except:
        article_author='None'
    new['作者']=article_author
    content=rows[0].find_all('article','container-fluid article')[0].find_all('span')
    new['新聞內文']=[(j.text.strip()) for j in content if 'class' not in j.attrs]              
    new['網址']=i
   # print(new)
    nn.append(new)
mo=pd.DataFrame.from_records(nn)
mo.to_excel('Taipei news.xlsx',index=False)
```
